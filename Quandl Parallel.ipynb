```python
import pandas as pd
import quandl as qd
import numpy as np
import multiprocessing as mp
%matplotlib inline
qd.ApiConfig.api_key = ""
```

```python
def get_data_from_quandl(ticker):
    try:
        res = qd.get("WIKI/{}".format(ticker))
    #     target[ticker] = res
        print(ticker)
        return res
    except:
        print(ticker, 'NaNa')
        return None
```

```python
sp500const = pd.read_csv('/Users//Projects/stocks/sp500_constituents.csv')
```

```python
from multiprocessing import Pool

tickers = sp500const.Symbol
pool = Pool(processes=5)
out = pool.map(get_data_from_quandl, tickers)
```
